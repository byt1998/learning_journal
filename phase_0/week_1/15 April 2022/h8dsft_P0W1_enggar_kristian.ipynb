**Dataset Description** \
\
This dataset includes 721 Pokemon, including their number, name, first and second type, and basic stats: HP, Attack, Defense, Special Attack, Special Defense, and Speed.

The feature of this data is:

- #: ID for each pokemon
- Name: Name of each pokemon
- Type 1: Each pokemon has a type, this determines weakness/resistance to attacks
- Type 2: Some pokemon are dual type and have 2
- Total: sum of all stats that come after this, a general guide to how strong a pokemon is
- HP: hit points, or health, defines how much damage a pokemon can withstand before fainting
- Attack: the base modifier for normal attacks (eg. Scratch, Punch)
- Defense: the base damage resistance against normal attacks
- SP Atk: special attack, the base modifier for special attacks (e.g. fire blast, bubble beam)
- SP Def: the base damage resistance against special attacks
- Speed: determines which pokemon attacks first each round

- Mengklasifikasikan Role pada karakter pokemon menjadi 3  yaitu: Attacker, Defender, Speedster, 

```python

pk_bahasa2.describe()
```

```python
pk_bahasa2["Tipe 1"].unique()
```

```python
# Statistik dari role attacker dengan Serangan lebih dari 103 atau lebih dari 75% nilainya 
# dan pertahanannya kurang dari 100 atau kurang dari 75%
pk_attacker = pk_bahasa2.query("Serangan >= 103 & Pertahanan <= 55")
pk_attacker
```

```python
print(pk_attacker.shape)
print("Terdapat 5 Pokemon Attacker")
```

```python
pk_attacker.groupby("Name")["Pertahanan Sp", "Pertahanan", "Serangan", "Kecepatan", "Serangan Sp"]\
    .sum().plot(kind="bar").set_title("Perbandingan Pokemon Attacker")
```

```python
# Statistik dari role speedster dengan Kecepatan lebih dari 91 atau lebih dari 75% nilainya 
# dan memiliki Pertahanan dan Pertahanan Spesial yang lemah kurang dari 25% atau <=55
pk_speedster = pk_bahasa2[
    (pk_bahasa2["Kecepatan"] >= 91) &
    (pk_bahasa2["Pertahanan"] <= 55) &
    (pk_bahasa2["Pertahanan Sp"] <= 55)
]
pk_speedster
```

```python
print(pk_speedster.shape)
print("Terdapat 7 Pokemon Speedster")
```

```python
pk_speedster.groupby("Name")["Pertahanan Sp", "Pertahanan", "Serangan", "Kecepatan", "Serangan Sp"]\
    .sum().plot(kind="bar").set_title("Perbandingan Pokemon Speedster")
```

```python
# Statistik dari role All Rounder dengan Serangan, Pertahanan, dan Kecepatan lebih dari 50% nilainya 
# dan memiliki yang lemah kurang dari 50% atau <= 75
pk_allrounder = pk_bahasa2[
    (pk_bahasa2["Serangan"] >= 80) &
    (pk_bahasa2["Pertahanan"] >= 75) &
    (pk_bahasa2["Kecepatan"] >= 75) &
    (pk_bahasa2["Pertahanan Sp"] <= 75)
]
pk_allrounder
```

```python
print(pk_allrounder.shape)
print("Terdapat 18 Pokemon All Rounder")
```

```python
pk_allrounder.groupby("Name")["Pertahanan Sp", "Pertahanan", "Serangan", "Kecepatan", "Serangan Sp"]\
    .sum().plot(kind="bar").set_title("Perbandingan Pokemon All Rounder")
```

```python
# Statistik dari role Defender dengan Pertahanan lebih dari 103 atau lebih dari 75%,
# Pertahanan Spesial lebih dari 75 atau lebih dari 50%
# dan Serangan kurang dari 59 atau kurang dari 25%
pk_defender = pk_bahasa2[
    (pk_bahasa2["Serangan"] <= 59) &
    (pk_bahasa2["Pertahanan"] >= 103) &
    (pk_bahasa2["Pertahanan Sp"] >= 75)
]
pk_defender
```

```python
print(pk_defender.shape)
print("Terdapat 8 Pokemon Defender")
```

```python
pk_defender.groupby("Name")["Pertahanan Sp", "Pertahanan", "Serangan", "Kecepatan", "Serangan Sp"]\
    .sum().plot(kind="bar").set_title("Perbandingan Pokemon Defender")
```

```python
# Statistik dari role Support dengan Pertahanan Sp lebih dari 75% nilainya 
# Pertahanan lebih dari 50%
# Serta Serangan, Kecepatan, dan Serangan Sp  kurang dari 50%
pk_support = pk_bahasa2[
    (pk_bahasa2["Pertahanan Sp"] >= 95) &
    (pk_bahasa2["Pertahanan"] >= 75) &
    (pk_bahasa2["Serangan"] <= 80) &
    (pk_bahasa2["Kecepatan"] <= 70) &
    (pk_bahasa2["Serangan Sp"] <= 70)
    
]
pk_support
```

```python
print(pk_support.shape)
print("Terdapat 6 Pokemon Support")
```

```python
pk_support.groupby("Name")["Pertahanan Sp", "Pertahanan", "Serangan", "Kecepatan", "Serangan Sp"]\
    .sum().plot(kind="bar").set_title("Perbandingan Pokemon Support")
```

# Conclusion

Kesimpulan yang didapat setelah dibagi menjadi beberapa role

```python
# Penggabungan Data Struktur pada pokemon Role
pk_role = pd.concat([pk_allrounder, pk_attacker, pk_defender, pk_speedster, pk_support], sort=False)
pk_role
```

```python
# Hasil 2D akhir baris dan kolom pada Pokemon Role
print(pk_role.shape)
print("Terdapat 44 baris dan 10 kolom hasil akhir")
```

```python
# Hasil akhir pada grafik pokemon
pk_role.groupby("Name")["Pertahanan Sp", "Pertahanan", "Serangan", "Kecepatan", "Serangan Sp"]\
    .sum().plot(kind="bar").set_title("Perbandingan Pokemon Role")
```
