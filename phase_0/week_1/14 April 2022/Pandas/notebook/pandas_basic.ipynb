## Columns Manipulattion
- create new columns
- rename column
- drop column

```python
# membuat salinan
df_new = df.copy()
```

```python
df_new.head()
```

```python
# create new column
df_new['selisih'] = df_new['pts'] - df_new['opp_pts']
```

```python
df_new.head()
```

```python
# jika tidak mau ada tanda minusnya tambah np.abs
df_new['selisih'] = np. abs(df_new['pts'] - df_new['opp_pts'])
```

```python
df_new.head()
```

rename column

```python
df_new.rename(columns={'game_location': 'loc', 
                        'game_result' : 'result'})
```

```python
# kolom belum berubah
df_new
```

```python
# merubahnya tambah variabel atau inplace=True
df_new.rename(columns={'game_location': 'loc', 
                        'game_result' : 'result'}, inplace=True)
```

```python
df_new
```

drop column

```python
# menghilangkan kolom elo_i dan elo_n
# di dalam default .drop adalah axis=0
df_new.drop(['elo_i', 'elo_n'], axis=1, inplace=True)
```

```python

df_new
```

drop rows
- drop year = 1947 & team spesifik = query
- index kita salin jadikan list
- list sebagai input .drop, axis = 0

```python
# creating new column based on exsisting information using loop
# siapkan list kosong
# buat program -> selisih pts dengan opp_pts per baris 
# simpan hasil perbaris ke list kosong
# convert list yang sudah terisi menjadi kolom
 
hasil = []

# df_new.iloc[:,10] - kolom pts
# df_new.iloc[:,14] - kolom opp_pts
for i in range(len(df_new)):
    selisih = df_new.iloc[i,10] - df_new.iloc[i,14]
    hasil.append(selisih)

df_new['menggunakan_loop'] = hasil
```
