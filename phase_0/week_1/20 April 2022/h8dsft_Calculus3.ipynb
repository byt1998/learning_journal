# Answer

fungsi awal 3ùë•+4ùë¶=5
maka diubah menjadi 
y = 5/4 - 3/4x atau y = 1.25 - 0.75x
x = 5/3 - 5/4y

```python
y = 1.25 - 0.75*x
x = 5/3 - 5/4*y
```

```python
# Define
X = np.linspace(-50,50)

# Define
Y = 1.25 - 0.75*X

result_gradient_x = np.gradient(Y)
result_gradient_x
```

```python
# Plot `x` and `result_gradient_x`

import matplotlib.pyplot as plt

fig, ax = plt.subplots(ncols=2, figsize=(21,5))

ax[0].plot(X, Y)
ax[0].set_title('Plot of `x`')

ax[1].plot(result_gradient_x, Y)
ax[1].set_title('Plot of `result_gradient_x`')
```

```python
# Define
Y = np.linspace(-50,50)

# Define
X = (5/3) - (5/4)*Y

result_gradient_y = np.gradient(X)
result_gradient_y
```

```python
# Plot `x` and `result_gradient_x`

import matplotlib.pyplot as plt

fig, ax = plt.subplots(ncols=2, figsize=(21,5))

ax[0].plot(X, Y)
ax[0].set_title('Plot of `x`')

ax[1].plot(result_gradient_y, X)
ax[1].set_title('Plot of `result_gradient_y`')
```
