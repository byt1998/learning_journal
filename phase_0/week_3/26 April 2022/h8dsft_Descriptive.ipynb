## 2. Skewness and Kurtosis

```python
import numpy as np
import scipy.stats as stats
```

```python
latitude = covid_19['New Cases'].sample(100)
density = stats.gaussian_kde(latitude)
n, x, _ = plt.hist(latitude, histtype='step', bins=25)  
plt.plot(x, density(x)*10**5)
plt.axvline(latitude.mean(), color='magenta', linestyle='dashed', linewidth=2)
plt.axvline(latitude.median(), color='green', linestyle='dashed', linewidth=2)
plt.show()
```

```python
covid_19.sample(5)
```

```python
# Get Skewness and Kurtosis Value
numcols = ['Total Cases', 'Total Deaths', 'Total Recovered']
for col in numcols:
    print(covid_19[col].name + ' skewness: ' + str(covid_19[col].skew()))
    print(covid_19[col].name + ' kurtosis: ' + str(covid_19[col].kurt()))
    density = stats.gaussian_kde(covid_19[col])
    n, x, _ = plt.hist(covid_19[col], histtype='step', bins=25)
    plt.plot(x, density(x)*6)
    plt.show()
    print('\n')
```

```python
# mendapatkan range dengan selisih .max() - .min()
numcols = ['Total Cases', 'Total Deaths', 'Total Recovered']
for col in numcols:
    print(covid_19[col].name + ' range: ' + str(covid_19[col].max() - covid_19[col].min()))
```

## 5. Varience and Standard Deviation

```python
# Get the Variance

print(covid_19['Total Cases'].var())
```

```python
# Get the Standard Deviation
print(covid_19['Total Cases'].std())
```

```python
# Create a random standard normal distribution
covid19 = pd.DataFrame(np.random.randn(25000, 1), columns=['Total Cases'])

# PLot the distribution as a histogram with a density curve

total_cases = covid19['Total Cases'].head(5)
density = stats.gaussian_kde(total_cases)
n, x, _ = plt.hist(total_cases, color='lightgrey', bins=100)
plt.plot(x, density(x))

# Get the mean and standard deviation
s = covid_19['Total Cases'].std()
m = covid_19['Total Cases'].mean()

# Annotate 1 stdev
x1 = [m-s, m+s]
y1 = [0.25, 0.25]
plt.plot(x1, y1, color='magenta')
plt.annotate('1S (68.26%)', (x1[1],y1[1]))

# Annotate 2 stdevs
x2 = [m-(s*2), m+(s*2)]
y2 = [0.05, 0.05]
plt.plot(x2,y2, color='green')
plt.annotate('2s (95.45%)', (x2[1],y2[1]))

# Annotate 3 stdevs
x3 = [m-(s*3), m+(s*3)]
y3 = [0.005, 0.005]
plt.plot(x2,y2, color='orange')
plt.annotate('2s (99.73%)', (x3[1],y3[1]))

# Show the location of the mean
plt.axvline(total_cases.mean(), color='grey', linestyle='dashed', linewidth=1)
plt.show()
```

```python
# Get Z-Score
covid_19['Z-Score'] = stats.zscore(covid_19['Total Cases'])
covid_19.head(5)
```
